```python
import pandas as pd
import os

# Specify the folder path
folder_path = 'D:\\Data_analyst\\cricket_insights\\t20_csv_files'

# Create an empty dictionary to store DataFrames
dfs = {}

# Loop through each file in the folder
for filename in os.listdir(folder_path):
    # Check if the file is a CSV file
    if filename.endswith('.csv'):
        # Construct the full file path
        file_path = os.path.join(folder_path, filename)
        
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        
        # Store the DataFrame in the dictionary with the filename as the key
        dfs[filename] = df
        df_name = os.path.splitext(filename)[0]
        
        # Store the DataFrame as a separate object
        globals()[df_name] = df

# Access the DataFrames using the filenames as keys
for filename, df in dfs.items():
    print(f"DataFrame for {filename}:")
    print(df)
    print("\n")
```

# Openers

```python
innings = batsman_info['name'].value_counts()
names_to_keep = innings[innings > 3].index
openers = batsman_info[batsman_info['name'].isin(names_to_keep)]
openers
```

```python
openers = openers.sort_values(by = 'name')
```

```python
openers = openers[openers['battingPos']<4].sort_values(by = 'name')
openers
```

```python
openers['SR'] = openers['SR'].astype(float)
```

```python
openers['boundary'] = openers['4s']*4+openers['6s']*6
openers
```

```python
openers.dtypes
```

```python
avg_SR_runs = openers.groupby('name')[['SR', 'runs', 'battingPos']].mean().reset_index()
avg_SR_runs.rename(columns={'runs': 'avg_runs', 'SR':'avg_SR'}, inplace=True)
avg_SR_runs['battingPos'] = avg_SR_runs['battingPos'].round().astype(int)
```

```python
boundary_perc = openers.groupby('name')[['boundary', 'runs']].sum().reset_index()
boundary_perc['boundary_percentage'] = boundary_perc['boundary']/boundary_perc['runs']*100
boundary_perc = boundary_perc[boundary_perc['boundary_percentage']>50].sort_values(by = 'name')
boundary_perc
```

```python
openers_eligible = pd.merge(avg_SR_runs, boundary_perc, on='name', how = 'inner')
openers_eligible
```

```python
openers_eligible = openers_eligible[openers_eligible['avg_runs']>30].sort_values(by = 'name')
openers_eligible
```

```python
openers_eligible.to_csv('openers.csv')
```

# middle order

```python
innings = batsman_info['name'].value_counts()
names_to_keep = innings[innings > 3].index
middle = batsman_info[batsman_info['name'].isin(names_to_keep)]
middle
```

```python
middle = middle.sort_values(by = 'name')
middle = middle[middle['battingPos']>2].sort_values(by = 'name')
middle
```

```python
avg_SR_runs_middle = middle.groupby('name')[['SR', 'runs', 'battingPos']].mean().reset_index()
avg_SR_runs_middle.rename(columns={'runs': 'avg_runs', 'SR':'avg_SR'}, inplace=True)
avg_SR_runs_middle['battingPos'] = avg_SR_runs_middle['battingPos'].round().astype(int)
avg_SR_runs_middle
```

```python
balls_middle = middle.groupby('name')[['balls']].sum().reset_index()
balls_middle
```

```python
middle_eligible = pd.merge(avg_SR_runs_middle, balls_middle, on='name', how = 'inner')
middle_eligible
```

```python
middle_eligible = middle_eligible[(middle_eligible['balls']>20) & (middle_eligible['avg_runs']>30) & (middle_eligible['avg_SR']>125)].sort_values(by = 'name')
middle_eligible
```

```python
middle_eligible.to_csv('middle.csv')
```

# All Rounders

```python
all_rounders_list = pd.merge(batsman_info, bowlers_info, on='name', how = 'inner')
innings_all = all_rounders_list['name'].value_counts()
names_to_keep_all = innings_all[innings_all > 2].index
all_round = all_rounders_list[all_rounders_list['name'].isin(names_to_keep)]
all_round
```

```python
all_round = all_round.sort_values(by = 'name')
all_round = all_round[all_round['battingPos']>4].sort_values(by = 'name')
all_round
```

```python
all_round['runs_x'] = all_round['runs_x'].astype(float)
all_round['SR'] = all_round['SR'].astype(float)
```

```python
avg_SR_runs_all_round = all_round.groupby('name')[['SR', 'runs_x', 'battingPos', 'runs_y', 'overs']].mean().reset_index()
avg_SR_runs_all_round.rename(columns={'runs_x': 'avg_runs', 'SR':'avg_SR'}, inplace=True)
avg_SR_runs_all_round['battingPos'] = avg_SR_runs_all_round['battingPos'].round().astype(int)
avg_SR_runs_all_round['runs/over'] = avg_SR_runs_all_round['runs_y']/avg_SR_runs_all_round['overs']
```

```python
avg_SR_runs_all_round
```

```python
balls_per_wicket = all_round.groupby('name')[['runs_y', 'wickets']].sum().reset_index()
balls_per_wicket['balls/wicket'] = balls_per_wicket['runs_y']/balls_per_wicket['wickets']
balls_per_wicket
```

```python
all_round_eligible = pd.merge(avg_SR_runs_all_round, balls_per_wicket, on = 'name', how = 'inner')
all_round_eligible
```

```python
all_round_eligible = all_round_eligible[(all_round_eligible['balls/wicket']<20) & (all_round_eligible['avg_runs']>15) & (all_round_eligible['runs/over']<7) & (all_round_eligible['avg_SR']>120)].sort_values(by = 'name')
all_round_eligible
```

```python
all_round_eligible.to_csv('all_rounders')
```

# Bowlers

```python
innings_bowler = bowlers_info['name'].value_counts()
names_to_keep_bowler = innings_bowler[innings_bowler > 4].index
bowlers = bowlers_info[bowlers_info['name'].isin(names_to_keep)]
bowlers = pd.merge(bowlers, player_info, on = 'name', how = 'inner')
bowlers
```

```python
bowlers = bowlers.sort_values(by = 'name')
fast = bowlers['bowlingStyle'].str.contains('fast', case=False, na=False)
bowlers = bowlers[fast]
```

```python
bowlers.drop(['team', 'image', 'description', 'playingRole'], axis = 1, inplace= True)
bowlers['balls'] = bowlers['overs']*6
bowlers
```

```python
balls_per_wicket_bowler = bowlers.groupby('name')[['balls', 'runs', 'wickets']].mean().reset_index()
balls_per_wicket_bowler['balls/wicket'] = balls_per_wicket_bowler['balls']/balls_per_wicket_bowler['wickets']
balls_per_wicket_bowler['runs/wicket'] = balls_per_wicket_bowler['runs']/balls_per_wicket_bowler['wickets']
balls_per_wicket_bowler
```

```python
dot_balls = bowlers.groupby('name')[['balls', '0s']].sum().reset_index()
dot_balls['dot_balls_perc'] = dot_balls['0s']/dot_balls['balls']*100
dot_balls
```

```python
bowlers_eligible = pd.merge(balls_per_wicket_bowler, dot_balls, on='name', how = 'inner')
bowlers_eligible
```

```python
bowlers_eligible = bowlers_eligible[(bowlers_eligible['balls/wicket']<16)& (bowlers_eligible['runs/wicket']<20) & (bowlers_eligible['dot_balls_perc']>40) ].sort_values(by = 'name')
bowlers_eligible
```

```python
bowlers_eligible.to_csv('bowlers')
```
